# Advertisement campaign analysis

Digital advertizing has risen to become the largest advertising channel. According to Statisca.com, the Global digital ad market is projected to reach USD 345 bilion in 2020. Two platforms, Facebook and Google at the forefront, accounting for almost 60% of the total global digital ad spending.

These platforms provide the capability to micro target different audiences and provide metrics to understand how the ad campaign is performing. Therefore, for this project work, we set out to analyse some advertizing metric dataset on ad campaigns ran on Google and Facebook to understand "What is the performance of various Ad runs given different target segments?"

## Data

Data source is from Kaggle https://www.kaggle.com/avinashlalith/merkle-sokrati-advertising-campaign
    
It is a dataset of Google and Facebook Marketing campaigns for the time period from October 2019 to July 2020 for some company. The dataset shows the performance of advertising campaigns for different Age-groups and dimensions.

The dataset includes about 16,800 rows and 16 features (columns). Google Ads accounts for 90% of the data while Facebook Ads account for the remaining 10%.

The features in the dataset:

- **Date**: This identifies the date on which the Marketing campaign was run.

- **Campaign Platform**: This identifies the platform where the marketing campaign has been deployed. There are two possible values, Google or Facebook Ads.

- **Device**: This identifies the medium through which the audience has consumed the Ad. Posible values are Desktop, Mobile, Tablet for Google Ads and Device for Facebook Ads.

- **Campaign type**: 90% of the dataset are Google Searches. The remaining 10% are Facebook Conversion campaigns.

- **Communication medium**: This typically refers to the trigger for the Ads. On Google Ads, the medium is specific search keywords. Facebook Ads medium are defined as Creative, which may include a number of different sub channels including visiting the company's Facebook page, brand/product search, etc.

- **SubChannel**: This is a feature of Google Ads. The subchannel allows the advertiser to set up ads based on specific words in the search keyword. Possible values include are Brand, Competitor, and Generic keywords.

- **Creative Name and type**: These are features of Facebook Ads. The Type define the Ad format e.g. Carousel (typically a scrollable video) or Image. The name on the other hand refers to the name given to the ad , Possible values are Girl, Click and Carousal

- **Age**: What age segment have been targetted by the markeing campaign. Example values are 18-24, 25-34, 35-44, 65 or over, etc.

- **Audience**: This is another feature of Facebook Ads. The feature refers to a grouping of the target audience in different campaigns (based on some undisclosed attributes). Possible values for Facebook Ad Audience is 1,2,3.

- **Spends**: This refers to the average spend by the target audience as a result of the advertising campaign.

- **Clicks**: This feature refers to the number of times the target group clicked on a specific ad in the Google- Ads and/or Facebook Ads campaign.

- **Impressions**: This feature refers to a count of the number of times a specific ad was shown to a member of a target group.

- **Link Clicks**: Facebook Ads feature, refers to the number of clicks on links within the ad that led to advertiser-specified destinations, on or off Facebook.

## Predictive Analysis

### Decision Trees

```python
# Import Decision Tree and Decision Tree Classifier
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
# Feature set
X = df2[['spends', 'age', 'impressions', 'clicks', 'link_clicks']]
# Target value
y = df['campaign_platform']

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

```python
X.head()
```

### Train Model

```python
# Create Decision Tree classifer object
# Entropy measures the level of impurity present in the data. 
clf = DecisionTreeClassifier(criterion="entropy", max_depth=5)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### Time Series with Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
# create new dataframe with only features we are interested in
df3 = df[['Date','clicks', 'spends']].copy()

#For time series analysis we need to assure that th index is in datetime format
df3['Date'] = pd.to_datetime(df3['Date'])
df3.index = df3['Date']
df3=df3.drop(['Date'], axis=1)
df3.head()
```

### Resample data and use weekly mean values

```python
#resample and use weekly mean values

a1=[df3['clicks'].resample('W').mean()]
a2= [df3['spends'].resample('W').mean()]

df_clicks=pd.DataFrame(a1)
df_clicks=df_clicks.transpose()

df_spends=pd.DataFrame(a2)
df_spends=df_spends.transpose()

df_mean = pd.concat([df_clicks, df_spends], axis=1)
```

```python
df_mean.head()
```

### Training a Single Exponential Smoothing model

```python
plt.figure(figsize=(12,6))
spend_fit = SimpleExpSmoothing(df_mean['spends']).fit()
spend_forecast = spend_fit.forecast(12)
spend_forecast.plot(label='forecasted spends')
spend_fit.fittedvalues.plot(label='observed spends')

click_fit = SimpleExpSmoothing(df_mean['clicks']).fit()
click_forecast = click_fit.forecast(12)

click_forecast.plot(label='forecasted clicks')
click_fit.fittedvalues.plot(label='observed clicks')

plt.legend()
plt.show()
```

```python
sqrt(mean_squared_error(df_mean['spends'], spend_fit.fittedvalues))
```

```python
mean_absolute_error(df_mean['spends'], spend_fit.fittedvalues)
```
